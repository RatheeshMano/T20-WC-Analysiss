## **To calculate the Bowling Strike Rate**

**Bowling Strike Rate= Total Wickets / TakenTotal Balls Bowled**

```python
valid_wickets= deliveries[deliveries['wicket_type'].notnull()]

#group by
ball_bowled = deliveries.groupby('bowler')['ball'].count()
wicket_taken = valid_wickets['bowler'].value_counts()

#bowling Strike rate
bowling_sr = (ball_bowled/wicket_taken).dropna()
bowling_sr = bowling_sr.sort_values()
bowling_sr
```

```python
bowling_sr.head(10).plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Top 10 Bowlers with Best Bowling Strike Rate')
plt.xlabel('Bowler')
plt.ylabel('Balls per Wicket')
plt.xticks(rotation=45)
plt.show()
```

```python
player_of_match_counts = matches['player_of_match'].value_counts()
print(player_of_match_counts.head(10))
```

## **Key Player for their Team**

```python
player_of_match_count = matches['player_of_match'].value_counts().reset_index()
player_of_match_count.columns = ['player', 'matches_won']
player_of_match_count = player_of_match_count.sort_values(by='matches_won', ascending=False)
print(player_of_match_count.head())  # Top players with most "Player of the Match" awards

winning_performers = matches[matches['winner'].notnull()]
winning_performers = winning_performers.groupby('player_of_match').agg({
    'winner_runs': 'sum',  # or 'winner_wickets' for bowler impact
    'winner_wickets': 'sum'
}).reset_index()
winning_performers = winning_performers.sort_values(by='winner_runs', ascending=False)
print(winning_performers.head())  # Players who performed best when their team won
```

```python
# Player of the Match Count (assuming player_of_match_count is calculated earlier)
sns.set(style="whitegrid")

# Plot the top players with most "Player of the Match" awards
plt.figure(figsize=(10, 6))
sns.barplot(x='matches_won', y='player', data=player_of_match_count.head(10), palette='viridis')
plt.title('Top Players with Most Player of the Match Awards')
plt.xlabel('Number of Matches Won')
plt.ylabel('Player')
plt.show()
```

## **Memorable Performances**

```python
large_margin_wins = matches[matches['winner_runs'].notnull() | matches['winner_wickets'].notnull()]
large_margin_wins = large_margin_wins.sort_values(by=['winner_runs', 'winner_wickets'], ascending=False)
print(large_margin_wins[['match_number', 'winner', 'winner_runs', 'winner_wickets']].head())

individual_performances = matches[['match_number', 'player_of_match', 'winner_runs', 'winner_wickets']]
individual_performances = individual_performances.dropna()
individual_performances = individual_performances.sort_values(by=['winner_runs', 'winner_wickets'], ascending=False)
print(individual_performances.head())
```

```python
# Large Margin Wins
plt.figure(figsize=(10, 6))
sns.barplot(x='winner_runs', y='match_number', data=large_margin_wins.head(10), palette='coolwarm')
plt.title('Top 10 Matches with Largest Winning Margins (by Runs)')
plt.xlabel('Winning Margin (Runs)')
plt.ylabel('Match Number')
plt.show()
```

## **Toss Impact: Win Probability**

```python
toss_impact = matches.groupby(['toss_winner', 'winner']).size().reset_index(name='match_count')
toss_impact = toss_impact.sort_values(by='match_count', ascending=False)
print(toss_impact.head())
```

```python
print(matches['winner'].unique())
print(matches['toss_winner'].unique())
matches['winner'] = matches['winner'].apply(lambda x: x.split(' ')[0] if isinstance(x, str) else x)
matches['toss_winner'] = matches['toss_winner'].apply(lambda x: x.split(' ')[0] if isinstance(x, str) else x)
```

```python
toss_outcome = toss_impact.groupby(['toss_winner', 'winner']).size().reset_index(name='match_count')
outcome_distribution = toss_outcome.groupby('winner')['match_count'].sum()

plt.figure(figsize=(10, 10))
plt.pie(outcome_distribution, labels=outcome_distribution.index, autopct='%1.1f%%', startangle=30, colors=sns.color_palette('Set3', n_colors=len(outcome_distribution)))
plt.title('Match Outcome Distribution Based on Toss Winner')
plt.show()
```

Based on the data and analysis, it has been observed that South Africa and India have shown the highest impact in determining match outcomes, particularly when it comes to their performance after winning the toss. Their consistent performance and strategic decisions post-toss have played a key role in propelling them to the finals in this particular tournament.

# **Conclusion:**

The consistency and performance of South Africa and India, driven by their decision-making post-toss, underline the significant influence of toss outcomes in this tournament. Their ability to capitalize on the early advantage has been a decisive factor in their journey to the finals, highlighting how important the toss can be in shaping the course of high-stakes matches.
